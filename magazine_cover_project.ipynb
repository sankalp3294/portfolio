# Project Overview
This notebook adds a magazine cover project card to the portfolio and makes the Photoshop-designed PNG available for checkout.

The goal is to load a local PNG file, display it inline, and include the project metadata so visitors can see the design immediately.

```python
from PIL import Image
from IPython.display import display

print('Libraries imported successfully.')
```

```python
cover_path = 'magazine-cover.png'
try:
    cover_image = Image.open(cover_path)
    print(f'Loaded image from {cover_path}')
except FileNotFoundError:
    print(f'File not found: {cover_path}')
    cover_image = None
```

```python
if cover_image:
    display(cover_image)
else:
    print('Cover image is not available to display.')
```

## Project Card Details

**Project Title:** Magazine Cover Design

**Description:** A Photoshop-designed magazine cover with bold editorial styling, sophisticated layout, and strong visual hierarchy.

**Tools Used:** Photoshop, PNG export

**Availability:** The magazine cover PNG is available in the project directory and can be downloaded directly.

```python
if cover_image:
    display(cover_image)
    cover_image.save('magazine-cover-checkout.png')
    print('Saved a copy as magazine-cover-checkout.png')
else:
    print('No image available to export.')
```
